# Reading the reponses 

```python
import re
import pandas as pd
import gspread
from oauth2client.service_account import ServiceAccountCredentials

def extract_numbers(text):
    return [int(x) for x in re.findall(r'\b\d+\b', text)]

# define the scope of the credentials and authenticate
creds = ServiceAccountCredentials.from_json_keyfile_name("/content/drive/MyDrive/ColabNotebooks/FinalProject/sheetconnection-377914-4b3a678b8b9e.json")
client = gspread.authorize(creds)
sheet = client.open("Data").sheet1

# initialize an empty pandas DataFrame
section_df = pd.DataFrame()

# populate the DataFrame with the data from the worksheet
for row in sheet.get_all_values():
    section_df = section_df.append(pd.Series(row), ignore_index=True)

# rename the columns of the DataFrame
section_df.columns = section_df.iloc[0]
section_df = section_df[1:]

# define the list of sections to extract numbers from
sections = ["SectionA", "SectionB", "SectionC", "SectionD", "SectionE", "SectionF", "SectionG"]

# apply the extract_numbers function to each section
for section in sections:
    section_df[section] = section_df[section].apply(extract_numbers)


print(section_df)
```

# Extracting and Validating

```python
def extract_answers(df, section_col, answers_col):
    num_questions = len(df[section_col].iloc[0])
    answer_ids = df[section_col].apply(lambda x: x[:num_questions]).values.tolist()[0]
    answers = df[answers_col].apply(lambda x: x.split(',')).values.tolist()[0]
    points = [int(answer.split('-')[1]) for answer in answers]
    answers = [int(answer.split('-')[0]) for answer in answers]
    if answer_ids == answers:
        answers_ok = True
    else:
        answers_ok = False
    points_ok = True if sum(points) == 10 else False
    return answer_ids, answers, points, answers_ok, points_ok

columns = ["SectionA", "AnswersA", "SectionB", "AnswersB","SectionC", "AnswersC","SectionD", "AnswersD","SectionE", "AnswersE","SectionF", "AnswersF","SectionG", "AnswersG"]

for i in range(0, len(columns), 2):
    section_col = columns[i]
    answers_col = columns[i+1]
    answer_ids, answers, points, answers_ok, points_ok = extract_answers(section_df, section_col, answers_col)
    print(f"Section {section_col[-1]} Answer {answers_col[-1]} IDs: {answer_ids}")
    print(f"Answers {answers_col[-1]} IDs: {answers}")
    print(f"Points {answers_col[-1]}: {points}")
    if answers_ok:
        print("Answers: OK")
    else:
        print("Answers: NOT OK")
    if points_ok:
        print("Points: OK")
    else:
        print("Points: NOT OK")


```
